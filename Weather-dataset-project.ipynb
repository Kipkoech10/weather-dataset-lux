```python
#Check if there are any NULL values present in the dataset.
df.isnull().sum()
```

```python
#Find all records where the weather was exactly clear.
clear_weather = df[df['Weather'] == 'Clear']
```

```python
with pd.option_context('display.max_rows', None,'display.max_columns',5):
    print(clear_weather)
```

```python
#Find the number of times the wind speed was exactly 4 km/hr.
wind_speed = df[df['Wind Speed_km/h'] == 4] 
count = len(wind_speed)
```

```python
print(f'The wind speed was exactly 4 km/hr {count} times.')
```

```python
#Rename the column "Weather" to "Weather_Condition."
df.rename(columns ={'Weather':'Weather_Condition'},inplace = True)
```

```python
df.head()
```

```python
#What is the mean visibility of the dataset?
df['Visibility_km'].mean()
```

```python
# Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
df[(df['Wind Speed_km/h'] > 24 ) & (df['Visibility_km'] == 25)]
```

```python
#What is the mean value of each column for each weather condition?
numeric_cols = df.select_dtypes(include=[np.number]).columns
mean_by_weather = df.groupby('Weather_Condition')[numeric_cols].mean()
print('\nmean values for each weather condition:')
print(mean_by_weather)
```

```python
#Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
df[((df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50)) | (df['Visibility_km'] > 40)]
```

```python
#Find the number of weather conditions that include snow.
snow_count = df['Weather_Condition'].str.contains('snow', case= False).sum()
print(snow_count)
```
